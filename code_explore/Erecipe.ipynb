```python
integer_columns = df.columns[6:]

temp = {column: int for column in integer_columns}
df = df.astype(temp)
df = df[(df['rating'] >= 0) & (df['rating'] <= 5)]
df = df[df['calories'] < 10000]
df = df[df['protein'] < 5000]
df = df[df['fat'] < 5000]
df = df[df['sodium'] < 5000]
df.dropna(inplace=True)
tags = df.columns[6:]
df['tags_count'] = df[tags].sum(axis=1)
df.drop(columns=['title'], axis=1, inplace=True)
print(df.head())
```

```python
def print_unique_and_nunique(df1):
    for column in df.columns:
        unique_values = df[column].unique()  # Get unique values in the column
        unique_count = df[column].nunique()  # Get count of unique values in the column
        print(f"Column '{column}':")
        print(f"  Unique values: {unique_values}")
        print(f"  Number of unique values: {unique_count}")
        print('-' * 50)
print_unique_and_nunique(df1)
```

```python
def remove_outliers_iqr(df1):
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1 = df1[column].quantile(0.25)
        Q3 = df1[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df1 = df1[(df1[column] >= lower_bound) & (df1[column] <= upper_bound)]
    return df1

df_no_outliers_iqr = remove_outliers_iqr(df1)
print("DataFrame after removing outliers using IQR:")
print(df_no_outliers_iqr)
```

```python
rate_tab = df1['rating'].value_counts().sort_index()
ratings = pd.DataFrame({
    'rating': rate_tab.index,
    'frequency': rate_tab.values
})
analysis_index = 1  # Replace with your analysis index
rh_title = f"Analysis {analysis_index}: Frequency of Ratings for Recipes"
rh_subtitle = (
    "Recipes ratings span discrete numeric values in Epicurious.\n"
    "Ratings is not a continuous variable."
)
rh_legend_title = "Rating"
rh_x_axis_label = "Rating"
rh_y_axis_label = "Volume of Recipes"

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(data=ratings, x='rating', y='frequency', palette='viridis', edgecolor='black')

plt.title(rh_title, fontsize=16)
plt.suptitle(rh_subtitle, fontsize=12)
plt.xlabel(rh_x_axis_label, fontsize=14)
plt.ylabel(rh_y_axis_label, fontsize=14)

plt.tight_layout()
plt.show()
```

**Skewed Rating Distribution:**

The ratings of recipes show a clear skew towards higher values, with most recipes receiving high ratings. The most frequent rating is 4.375, applied to nearly 6,000 recipes. Overall, the majority of recipes have ratings of 3.75 or higher, indicating a bias towards positive reviews.

**U-Shaped Pattern:**

The distribution also follows a notable U-shape, with peaks at both ends of the scale. A large number of recipes have perfect 5.0 ratings (around 1,800), while a significant amount also have 0.0 ratings (approximately 1,200).
